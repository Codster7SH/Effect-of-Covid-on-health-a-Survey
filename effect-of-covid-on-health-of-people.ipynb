**Effect on Covid on health of People a Qualitative and Quantitative analysis**

```python
import pandas as pd
import numpy as np
```

```python
df=('/kaggle/working/health_data.csv')
```

```python
df
```

```python
# Descriptive statistics for age
age_stats = df['Age'].describe()

# Count of chronic disease cases pre and post pandemic
chronic_disease_counts = df[['Pre_Chronic_Disease', 'Post_Chronic_Disease']].apply(pd.Series.value_counts)

# Mean and standard deviation of BMI pre and post pandemic
bmi_stats = df[['Pre_BMI', 'Post_BMI']].describe()

# Count of physical activity levels pre and post pandemic
physical_activity_counts = df[['Pre_Physical_Activity', 'Post_Physical_Activity']].apply(pd.Series.value_counts)

# Count of mental health status pre and post pandemic
mental_health_counts = df[['Pre_Mental_Health', 'Post_Mental_Health']].apply(pd.Series.value_counts)

print("Age Statistics:\n", age_stats)
print("\nChronic Disease Counts:\n", chronic_disease_counts)
print("\nBMI Statistics:\n", bmi_stats)
print("\nPhysical Activity Counts:\n", physical_activity_counts)
print("\nMental Health Counts:\n", mental_health_counts)
```

**Interpretation:**

This summary provides an overview of the age distribution in our dataset:

The average age of participants is around 28.6 years.
The standard deviation is approximately 4.02 years, indicating some variability in age.
Ages range from 22 to 35 years, mostly covering the youths.


* For Chronic disease, we deduce that -

1. Before the pandemic,around 21% of individuals reported having a chronic disease, while aprrox 79% did not.
2. After the pandemic, 24% reported having a chronic disease, indicating an increase in chronic disease incidence.

* For Mental Health, we deduce that -
   
1. Before the pandemic, 72% of individuals reported having a good mental health, while 18% reported it to be moderate, while only 9% reported it to be poor.
2. After the pandemic, with the sharp downfall we see 48% of individuals reported having a good mental health, while 30% reported it to be moderate, while only 21% reported it to be poor. which clearly indicates that mental health of people is highly affected due to the pandemic.

* For Physical Activity, we deduce that -

1. Before covid 31% of people were involved in high physical activity routine, while 39% people reported it to be Moderate and 29% had a routine of low activity.
2. while after the covid 19% of people were involved in high physical activity routine, while 39% people reported it to be Moderate and 40% had a routine of low activity. clearly indicates the effect of covid on people's physical activity levels.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization of Chronic Disease Counts
chronic_disease_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Chronic Disease Pre and Post Pandemic')
plt.ylabel('Count')
plt.show()

# Visualization of BMI Distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Pre_BMI'], color='blue', label='Pre-Pandemic BMI', kde=True)
sns.histplot(df['Post_BMI'], color='red', label='Post-Pandemic BMI', kde=True)
plt.title('BMI Distribution Pre and Post Pandemic')
plt.legend()
plt.show()

# Visualization of Physical Activity Levels
physical_activity_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Physical Activity Levels Pre and Post Pandemic')
plt.ylabel('Count')
plt.show()

# Visualization of Mental Health Status
mental_health_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Mental Health Status Pre and Post Pandemic')
plt.ylabel('Count')
plt.show()
```

```python
from scipy.stats import ttest_ind, chi2_contingency

# T-test for BMI pre and post pandemic
t_stat, p_val_bmi = ttest_ind(df['Pre_BMI'], df['Post_BMI'])
print(f'T-test for BMI: t-statistic = {t_stat}, p-value = {p_val_bmi}')

# Chi-square test for Chronic Disease
chi2, p_val_chronic, _, _ = chi2_contingency(chronic_disease_counts)
print(f'Chi-square test for Chronic Disease: chi2 = {chi2}, p-value = {p_val_chronic}')

# Chi-square test for Physical Activity Levels
chi2, p_val_activity, _, _ = chi2_contingency(physical_activity_counts)
print(f'Chi-square test for Physical Activity Levels: chi2 = {chi2}, p-value = {p_val_activity}')

# Chi-square test for Mental Health Status
chi2, p_val_mental_health, _, _ = chi2_contingency(mental_health_counts)
print(f'Chi-square test for Mental Health Status: chi2 = {chi2}, p-value = {p_val_mental_health}')
```

* Chi-square test for Chronic Disease: chi2 = 1.6387495, p-value = 0.20049
* Chi-square test for Physical Activity Levels: chi2 = 45.1580, p-value = 1.56334
* Chi-square test for Mental Health Status: chi2 = 118.65, p-value < 1.712959

```python
# Filter data for high and low physical activity levels
high_activity_pre = df[df['Pre_Physical_Activity'] == 'High']['Age'].value_counts().sort_index()
low_activity_pre = df[df['Pre_Physical_Activity'] == 'Low']['Age'].value_counts().sort_index()

high_activity_post = df[df['Post_Physical_Activity'] == 'High']['Age'].value_counts().sort_index()
low_activity_post = df[df['Post_Physical_Activity'] == 'Low']['Age'].value_counts().sort_index()

# Plot line graphs
plt.figure(figsize=(10, 6))

# High Physical Activity Comparison
plt.plot(high_activity_pre.index, high_activity_pre.values, label='High Activity Pre-Pandemic', marker='o')
plt.plot(high_activity_post.index, high_activity_post.values, label='High Activity Post-Pandemic', marker='o')

# Low Physical Activity Comparison
plt.plot(low_activity_pre.index, low_activity_pre.values, label='Low Activity Pre-Pandemic', marker='o')
plt.plot(low_activity_post.index, low_activity_post.values, label='Low Activity Post-Pandemic', marker='o')

plt.title('Physical Activity Levels Comparison')
plt.xlabel('Age')
plt.ylabel('Count%')
plt.legend()
plt.grid(True)
plt.show()
```
